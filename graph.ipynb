```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sqlite3
from matplotlib.ticker import MaxNLocator

conn = sqlite3.connect('pesca.db')
cursor = conn.cursor()
```

```python
# Carica i dati dalla tabella partecipazione_totale_nazionale in un DataFrame
df = pd.read_sql_query("SELECT * FROM occupazione_percentuale_nazionale ORDER BY Anno", conn)

#per DataWrangler si usa head()
#df.head()

plt.figure(figsize=(10, 6))

plt.plot(df['Anno'], df['Percentuale'], marker='o', linestyle='-', color='black')

plt.title('Occupazione percentuale nazionale')
plt.xlabel('Anno')
plt.ylabel('Occupazione Percentuale')

plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

# Mostra griglia per maggiore leggibilità
plt.grid(True)

# Mostra il grafico
plt.show()
```

```python
df = pd.read_sql_query("SELECT * FROM occupazione_percentuale_nazionale ORDER BY Anno", conn)

plt.figure(figsize=(10, 6))

plt.bar(df['Anno'], df['Percentuale'],  color="purple", alpha=0.8, label="dati")

# definisce l'inclinzaione delle labels su
plt.xticks(rotation=45)

plt.title("Occupazione percentuale nazionale")
plt.show()
```

```python
df = pd.read_sql_query("SELECT * FROM occupazione_percentuale_aree ORDER BY Anno", conn)

df_pivot = df.pivot(index='Anno', columns='Area_Geografica', values='Percentuale')

anni = df_pivot.index.tolist()       
aree = df_pivot.columns.tolist()    
x = np.arange(len(anni))         
bar_width = 0.8 / len(aree)         

plt.figure(figsize=(12, 6))

for i, area in enumerate(aree):
    valori = df_pivot[area].values  
    plt.bar(x + i * bar_width, valori, width=bar_width, label=area, alpha=0.85)

plt.xticks(x + bar_width * (len(aree)-1)/2, anni)
plt.xlabel("Anno")
plt.ylabel("Percentuale divisa per area")
plt.title("Occupazione percentuale divisa per anno e area geografica")
plt.legend(title="Area Geografica")
plt.tight_layout()
plt.show()
```
